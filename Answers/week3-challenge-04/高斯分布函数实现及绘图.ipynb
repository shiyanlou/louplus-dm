<font color='red'><i class="fa fa-check-square" aria-hidden="true"> 挑战</i></font>：参考高斯分布公式，使用 Python 实现高斯分布函数。

```python
"""实现高斯分布函数
"""
import numpy as np

def Gaussian(x, u, d):
    """
    参数:
    x -- 变量
    u -- 均值
    d -- 标准差

    返回:
    p -- 高斯分布值
    """
    ### 代码开始 ### (≈ 3~5 行代码)
    d_2 = d * d * 2
    zhishu = -(np.square(x - u) / d_2)
    exp = np.exp(zhishu)
    pi = np.pi
    xishu = 1 / (np.sqrt(2 * pi) * d)
    p = xishu * exp
    return p
    ### 代码结束 ###
```

<font color='red'><i class="fa fa-check-square" aria-hidden="true"> 挑战</i></font>：按规定的参数绘制高斯分布图像。

```python
from matplotlib import pyplot as plt
%matplotlib inline

### 代码开始 ### (≈ 5~10 行代码)
y_1 = Gaussian(x, 0, 1)
y_2 = Gaussian(x, -1, 2)
y_3 = Gaussian(x, 1, 0.5)
y_4 = Gaussian(x, 0.5, 5)

plt.figure(figsize=(8,5))
plt.plot(x, y_1, c='r', label="u=0, d=1")
plt.plot(x, y_2, c='b', label="u=-1, d=2")
plt.plot(x, y_3, c='g', label="u=1, d=0.5")
plt.plot(x, y_4, c='y', label="u=0.5, d=5")
plt.legend()
### 代码结束 ###
```
